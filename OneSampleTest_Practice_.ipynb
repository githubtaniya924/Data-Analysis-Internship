## Scale data

```python
import pandas as pd
import scipy
gss_data=pd.read_csv('GSS2012a.csv')

scalevar=gss_data['age'].dropna()
scalevar=scalevar.replace({'89 OR OLDER':90})
scalevar=pd.to_numeric(scalevar)
```

```python
'''
RQ - The average age is 45

H0 -> The average age is not different it is same as 45
Ha -> The average age is different

To find whether h0 is rejected or not we need to do one sample student_t test

if pvalue < 0.05
null hypothesis is rejected
else: null hypothesis is not rejected

'''
```

```python
from scipy.stats import ttest_1samp
ttest_1samp(a=scalevar,popmean=45)
```

```python
'''
Here the null hypothesis is rejected since pvalue < 0.05

The average age is different from 45.

now we can find the effect size using cohen d

'''
```

```python
#ha -> The mean is differnt from the hypothetical mean
#hypothetical mean is the assumed mean
#find effect size

m = scalevar.mean()
sd = scalevar.std()

mean=45

temp = (m -mean)/sd
cohen_d = abs(temp * 2**0.5)

cohen_d

if(cohen_d < 0.2):
  print('Negligible')
elif(cohen_d < 0.5):
  print('small')
elif(cohen_d <0.8):
  print('medium')
else:
  print('large')


```

## Hypothesis test on Ordinal data

```python
gss_data1=pd.read_csv('GSS2012a.csv')
ordinal_data=gss_data1['accntsci']
ordinal_data.value_counts().index
```

```python
ordinal_data = ordinal_data.replace({'Not too scientific' :0,
                                 'Not scientific at all' :1,
                                 'Pretty scientific' :2,
                                 'Very scientific' :3

})
```

```python
from scipy.stats import wilcoxon
ordinal_data=ordinal_data.dropna()
m=2.5
wilcoxon(x=ordinal_data-2.5,zero_method="wilcox")
```

```python
pvalue=8.37e-142

#pvalue < 0.05 therefore the null hypothesis is rejected
# now we need to find the effect size also known as rosenthal correlation coefficient
from scipy.stats import norm
z= norm.ppf(pvalue/2)

n = len(ordinal_data)
n

r_coff = abs( z / n ** 0.5)

if r_coff <0.2 :
  print('negligible')
elif r_coff <0.5 :
  print('small')
elif r_coff <0.8:
  print('medium')
else:
  print('large')

```
