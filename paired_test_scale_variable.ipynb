## Scale Variable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True, as_frame=True)

X.head()
y.head()
y.unique()
```

```python
data = pd.concat([X,y],axis=1)
data.head()
```

```python
data['target']= data['target'].replace({0:'type_1',1:'type_2',2:'type_3'})
data.head()
```

```python
data = data.rename(columns={
    'sepal length (cm)':'sepal_length',
    'sepal width (cm)':'sepal_width',
    'petal length (cm)':'petal_length',
    'petal width (cm)':'petal_width'
})
data.head()
```

```python
from statsmodels.multivariate.manova import MANOVA

result = MANOVA.from_formula('sepal_length+sepal_width+petal_length+petal_width~target',data=data)
```

```python
print(result.mv_test())
```

```python
pvalue = 0.0000
if(pvalue < 0.05):
  print('null hypothesis is rejected. Ha -> There is significant difference')
else:
  print('Null hypothesis cannot be rejected.The difference is not significant')
```

## Ordinal Variables

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data1= pd.read_csv('StudentStatistics.csv', sep=';')
data1.head()
```

```python
data1 = data1[['Teach_Motivate','Teach_LinkTheory','Teach_StimAsk','Teach_Avail']]
```

```python
data1.head()
```

```python
coding = {'Fully Disagree':0,'Disagree':1,'Neither disagree nor agree':2,'Agree':3,'Fully agree':5}
data1.replace(coding,inplace=True)
data1.head()
```

```python
data1.dropna(inplace=True)
from scipy.stats import friedmanchisquare
friedmanchisquare(data1['Teach_Motivate'],data1['Teach_LinkTheory'],data1['Teach_StimAsk'],data1['Teach_Avail'])
```

```python
pvalue=2.1220432262951938e-07
if(pvalue <0.05):
  print('null hypothesis is rejected. Ha -> There is significant difference')
else:
  print('Null hypothesis cannot be rejected.The difference is not significant')
```

```python
statistic=33.85869565217378
pvalue=2.1220432262951938e-07

#effect size -> kendall W
n=data1.shape[0]
k=data1.shape[1]

kendall = statistic / (n*(k-1))
kendall
```

```python
if(kendall < 0.3):
  print('weak')
elif(kendall < 0.5):
  print('medium')
else:
  print('strong')
```

```python
#The different variables were not having the same responses and after friedmanchisquare test
#it is found that the difference is significant
#The effect size according to kendall W test is weak
```
